### Question 1 (part 1)
- Crawl pages whose seed url is the press releases page of the Fed; 
- collect the word "covid" found within the page. Collect at least 10 such urls.
- the code should output the urls of the webpages found to contain the word "covid". 


```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import urllib.request
```

```python
seed_url = "https://www.federalreserve.gov/newsevents/pressreleases.htm"
req_fed = urllib.request.Request(seed_url, headers = {'User-Agent':'Mozilla/5.0'})
webpage = urllib.request.urlopen(req_fed).read()
```

#### Get all the child links from this seed url

```python
soup = BeautifulSoup(webpage)
links = []
for tag in soup.find_all('a', href = True): #find tags with links
    child = tag['href']
    links.append(urllib.parse.urljoin(seed_url, child))
links
```

### This is the set of links from the seed_url that contains word "covid":

```python
results = []
for l in list(set(links))[:200]:  # only visit the first 300 links, skip the ones that denies my request
    req = urllib.request.Request(l, headers = {'User-Agent':'Mozilla/5.0'})
    try:
        read_req = urllib.request.urlopen(req).read().lower() #lower the texts in this url
        if b'covid' in read_req:
            print(l)
            print("This is the "+str(list(set(links)).index(l))+" link in total")
            results.append(l)
    except:
        print("Request denied")
    
```

```python
print("I collected "+str(len(results))+" urls that contains word 'covid'")
```

### Question 1 (part 2): crawl pages whose seed url is the press releases pages of the SEC and collect urls of press releases that contain the word 'charges'.
The code should output the first 20 such links. For each link output the url and the text.

```python
sec_seed = "https://www.sec.gov/news/pressreleases"
req_sec = urllib.request.Request(sec_seed, headers = {'User-Agent':'Mozilla/5.0'})
webpage_sec = urllib.request.urlopen(req_sec).read()
```

```python
soup_sec = BeautifulSoup(webpage_sec)
links_sec = []
for tag in soup_sec.find_all('a', href = True): #find tags with links
    child = tag['href']
    links_sec.append(urllib.parse.urljoin(sec_seed, child))
```

```python
final_links = list(set(links_sec))
```

#### Here I print:
- the url that contains 'charges'
- the order of this url from the seed_url
- the texts that contains 'charges' in this url

```python
results_sec = []
for l in final_links[:50]:  # only visit the first 50 links, skip the ones that denies my request
    req = urllib.request.Request(l, headers = {'User-Agent':'Mozilla/5.0'})
    try:
        read_req = urllib.request.urlopen(req).read().lower() #lower the texts in this url
        
        if b'charges' in read_req:
            print(l)
            soup = BeautifulSoup(read_req, 'html.parser')
            texts = soup.get_text().split("\n")
            print("This is the "+str(final_links.index(l))+" link in total")
            print([u for u in texts if 'charges' in u])
            results_sec.append(l)
            
    except:
        print("Request denied")
    
```

```python
print("I collected "+str(len(results_sec))+" urls that contains word 'charges'")
```
